# Decision Tree

### Wine

```python
from sklearn.datasets import load_wine  # dataset 불러오기
from sklearn.model_selection import train_test_split  # train 과 test 나누기
from sklearn.tree import DecisionTreeClassifier  # Tree 분류기
```

#### 모델 생성 및 확인

```python
wine_data = load_wine()

X_train, X_test, y_train, y_test = train_test_split(wine_data.data, wine_data.target, test_size=0.2, random_state=10)

tree_clf = DecisionTreeClassifier(max_depth=3)
tree_clf.fit(X_train, y_train)

print("훈련 데이터 점수:", round(tree_clf.score(X_train, y_train), 3))
print("테스트 데이터 점수:", round(tree_clf.score(X_test, y_test), 3))
```

#### 시각화

```python
import matplotlib.pyplot as plt
import numpy as np

n_features = X_train.shape[1]
plt.barh(np.arange(n_features), tree_clf.feature_importances_, align="center")
plt.yticks(np.arange(n_features), wine_data.feature_names, fontsize=20)

plt.rcParams["figure.figsize"] = (10,10) #바그래프 크기 조정
plt.title('wine Classifier', fontsize=30) #제목 글자 크기 30
plt.xlabel('Importance level (%)', fontsize=20) #X 제목 글자 크기 20
plt.ylabel('Feature Variables', fontsize=10) #Y 제목 글자 크기 20
```
